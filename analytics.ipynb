```python
from main import getPuuid, getMatchHistory, playersTimeline
import matplotlib.pyplot as plt
```

```python
#function that takes in puuid and number(def: 10) outputs a graph of your gold over duration
def goldAnalysis(puuid, num=10):
    Matches = getMatchHistory(puuid)
    i = 1
    for game in Matches:
        if i <= num:
            gold, time = playersTimeline(game, puuid)
            plt.plot(time, gold)
        else:
            break
        i += 1

    plt.xlabel("Game time (minutes)")
    plt.ylabel("Total gold")
    plt.title("Gold over time")
    plt.grid(True)
    plt.show()

# Example usage:
# puuid = getPuuid("Zignov", "3644")
# goldAnalysis(puuid, num=10)
```

```python
puuid = getPuuid("Zignov", "3644")
#puuid = getPuuid("LivelyHarsh", "3145")
goldAnalysis(puuid, 15)
```
